```python
# Load my data
df = pd.read_csv('Favorite_music_survery.csv')
df.rename(columns={'Which year was your favorite song released?': 'Song Release Year',
                   'What is the genre of the music?': 'Song Genre',
                  'Which age group do you fall under?': 'Age Group',
                  'How often do you listen to music' : 'Song Frequency'}, inplace=True)
df
```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
```

```python
# extract the first year from a range or leave it if it's a single year
def clean_release_year(year):
    if isinstance(year, str):  # Check if it's a string
        year = year.split('-')[0].strip()  # Extract the first year from ranges
    return pd.to_numeric(year, errors='coerce')  # Convert to numeric
```

```python
# copy dataframe before modifying
df = df.copy() 

# Modify dataset by applying the function to clean the 'Song Release Year' column
df['Song Release Year'] = df['Song Release Year'].apply(clean_release_year)

# Drop any remaining NaN values in 'Song Release Year'
df.dropna(subset=['Song Release Year'], inplace=True) 
```

```python
# Focus on key columns
df = df[['Age Group', 'Song Genre', 'Song Release Year', 'Favourite Song']]
df
```

```python
# Encode categorical data (age group, genre)
# Combine age group and genre into a single feature
df['Age_Genre'] = df['Age Group'] + "_" + df['Song Genre']
# df['Age_Genre'] = df['Age Group'].astype(str) + "_" + df['Song Genre'].astype(str)
# df.dropna(subset=['Age_Genre'], inplace=True)

# One-hot encode Age_Genre
encoder = OneHotEncoder(sparse_output=False)
age_genre_encoded = encoder.fit_transform(df[['Age_Genre']])
```

```python
# Scale release year (0-1 range)
scaler = MinMaxScaler()
release_year_scaled = scaler.fit_transform(df[['Song Release Year']])
```

```python
# Combine all features
features = pd.concat([
    pd.DataFrame(age_genre_encoded, columns=encoder.get_feature_names_out(['Age_Genre'])),
    pd.DataFrame(release_year_scaled, columns=['Release Year'])
], axis=1)

df
```

```python
# Training the K-Nearest Neighbors Model
# Use KNN to find songs with similar age-genre-year patterns

knn = NearestNeighbors(n_neighbors=5, metric='cosine')
knn.fit(features)

def recommend_songs(age_group, genre, release_year):
    # Preprocess user input
    user_age_genre = f"{age_group}_{genre}"
    user_input_encoded = encoder.transform([[user_age_genre]])
    user_year_scaled = scaler.transform([[release_year]])
    
    # Combine features
    user_features = pd.concat([
        pd.DataFrame(user_input_encoded, columns=encoder.get_feature_names_out(['Age_Genre'])),
        pd.DataFrame(user_year_scaled, columns=['Release Year'])
    ], axis=1)
    
    # Find similar songs
    distances, indices = knn.kneighbors(user_features)
    recommendations = df.iloc[indices[0]]['Favourite Song'].tolist()
    return recommendations
```

```python
# Testing the model
# User input
age = "Generation Z  (13 Years - 28 Years)"
genre = "Afrobeats"
year = 2025  # User prefers songs around this year

# age_2 = "Millennials (29 Years - 44 Years)"
# genre_2 = "Hip-Hop/Rap"
# year_2 = 2000  # User prefers songs around this year


# Get recommendations
recommended_songs = recommend_songs(age, genre, year)
print("Recommended Songs:")
for song in recommended_songs:
    print(f"- {song}")


```
